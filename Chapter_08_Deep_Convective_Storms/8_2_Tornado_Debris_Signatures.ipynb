# 8.2 龙卷碎片特征 (TDS)

**参考文献**：Ryzhkov & Zrnic, Chapter 8, pp. 801-830

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def plot_tds_signature():
    """TDS特征示意"""
    echo_types = ['普通降水', '龙卷碎片 (TDS)']
    
    # 典型值
    Z_vals = [40, 35]
    ZDR_vals = [2.0, 0.5]
    rho_hv_vals = [0.95, 0.6]
    LDR_vals = [-20, -8]
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    ax1 = axes[0]
    bars = ax1.bar(echo_types, rho_hv_vals, color=['blue', 'red'], alpha=0.7)
    ax1.axhline(y=0.8, color='orange', linestyle='--', label='TDS阈值')
    ax1.set_ylabel('|ρhv|', fontsize=11)
    ax1.set_title('相关系数对比', fontsize=12)
    ax1.legend()
    for bar, val in zip(bars, rho_hv_vals):
        ax1.text(bar.get_x() + bar.get_width()/2, val + 0.02, f'{val:.2f}', ha='center')
    
    ax2 = axes[1]
    bars = ax2.bar(echo_types, LDR_vals, color=['blue', 'red'], alpha=0.7)
    ax2.set_ylabel('LDR (dB)', fontsize=11)
    ax2.set_title('退极化比对比', fontsize=12)
    
    ax3 = axes[2]
    bars = ax3.bar(echo_types, Z_vals, color=['blue', 'red'], alpha=0.7)
    ax3.set_ylabel('Z (dBZ)', fontsize=11)
    ax3.set_title('反射率对比', fontsize=12)
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== TDS核心特征 ===")
    print("ρhv 显著降低 (<0.8)")
    print("LDR 升高 (> -12 dB)")
    print("Z 适中 (非最大)")
    print("这些特征由碎片的多样性和随机取向导致")
```

```python
plot_tds_signature()
```
