# 8.3 ZDR柱状结构

**参考文献**：Ryzhkov & Zrnic, Chapter 8, pp. 831-860

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def plot_zdr_arch():
    """ZDR柱状结构示意"""
    height = np.linspace(0, 10, 100)
    
    # ZDR柱：上升气流区大雨滴被抬升
    ZDR = np.zeros_like(height)
    ZDR[height < 3] = 1.5
    ZDR[(height >= 3) & (height < 6)] = 3.5 - 0.5 * (height[(height >= 3) & (height < 6)] - 3)
    ZDR[height >= 6] = 1.0 + 0.1 * height[height >= 6]
    
    # 风暴主体
    Z = np.zeros_like(height)
    Z[height < 2] = 25 + height[height < 2] * 3
    Z[(height >= 2) & (height < 5)] = 35 - 2 * (height[(height >= 2) & (height < 5)] - 2)
    Z[height >= 5] = 30 + 0.5 * height[height >= 5]
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    ax1 = axes[0]
    ax1.plot(ZDR, height, 'g-', linewidth=2)
    ax1.fill_betweenx(height, ZDR, alpha=0.3)
    ax1.axhline(y=3, color='red', linestyle='--', label='ZDR柱起始')
    ax1.annotate('ZDR柱', xy=(3.5, 4), fontsize=11)
    ax1.set_xlabel('ZDR (dB)', fontsize=11)
    ax1.set_ylabel('高度 (km)', fontsize=11)
    ax1.set_title('ZDR柱状结构 - 上升气流指示', fontsize=12)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    ax2 = axes[1]
    ax2.plot(Z, height, 'b-', linewidth=2)
    ax2.set_xlabel('Z (dBZ)', fontsize=11)
    ax2.set_ylabel('高度 (km)', fontsize=11)
    ax2.set_title('反射率廓线', fontsize=12)
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== ZDR柱特征 ===")
    print("位置：风暴中低层（2-5km）")
    print("特征：ZDR峰值 > 3 dB")
    print("物理：上升气流将大雨滴抬升至高空")
    print("应用：定位强上升气流区，评估对流强度")
```

```python
plot_zdr_arch()
```
