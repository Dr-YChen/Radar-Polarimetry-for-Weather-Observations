# 8.1 冰雹信号识别

**学习目标**
- 理解冰雹的极化特征
- 掌握冰雹识别算法

**参考文献**：Ryzhkov & Zrnic, Chapter 8, pp. 761-800

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def hail_detection(Z, ZDR, KDP, rho_hv):
    """
    冰雹检测算法
    基于多参数阈值判断
    """
    # 冰雹指标
    hail_prob = 0.0
    reasons = []
    
    # 高ZDR是冰雹信号
    if ZDR > 3.5:
        hail_prob += 0.3
        reasons.append('高ZDR')
    
    # 低ρhv（形状不规则）
    if rho_hv < 0.85:
        hail_prob += 0.3
        reasons.append('低ρhv')
    
    # 高KDP（大粒子）
    if KDP > 2:
        hail_prob += 0.2
        reasons.append('高KDP')
    
    # 高Z（强反射）
    if Z > 50:
        hail_prob += 0.2
        reasons.append('高Z')
    
    is_hail = hail_prob >= 0.6
    
    return is_hail, hail_prob, reasons

def plot_hail_detection():
    """冰雹检测示意"""
    # 模拟数据点
    n = 200
    Z = np.random.randn(n) * 10 + 40
    ZDR = np.random.randn(n) * 1.5 + 2
    KDP = np.random.randn(n) * 0.8 + 1
    rho_hv = np.random.randn(n) * 0.1 + 0.92
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # 散点图
    ax1 = axes[0, 0]
    colors = []
    for i in range(n):
        is_hail, prob, _ = hail_detection(Z[i], ZDR[i], KDP[i], rho_hv[i])
        colors.append('red' if is_hail else 'blue')
    ax1.scatter(Z, ZDR, c=colors, s=30, alpha=0.6)
    ax1.axhline(y=3.5, color='red', linestyle='--', label='ZDR阈值')
    ax1.axvline(x=50, color='orange', linestyle='--', label='Z阈值')
    ax1.set_xlabel('Z (dBZ)', fontsize=11)
    ax1.set_ylabel('ZDR (dB)', fontsize=11)
    ax1.set_title('冰雹检测 (Z vs ZDR)', fontsize=12)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Z vs ρhv
    ax2 = axes[0, 1]
    ax2.scatter(Z, rho_hv, c=colors, s=30, alpha=0.6)
    ax2.axhline(y=0.85, color='red', linestyle='--', label='ρhv阈值')
    ax2.set_xlabel('Z (dBZ)', fontsize=11)
    ax2.set_ylabel('|ρhv|', fontsize=11)
    ax2.set_title('冰雹检测 (Z vs ρhv)', fontsize=12)
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 冰雹概率分布
    ax3 = axes[1, 0]
    probs = []
    for i in range(n):
        _, prob, _ = hail_detection(Z[i], ZDR[i], KDP[i], rho_hv[i])
        probs.append(prob)
    ax3.hist(probs, bins=20, color='coral', alpha=0.7, edgecolor='black')
    ax3.axvline(x=0.6, color='red', linestyle='--', label='检测阈值')
    ax3.set_xlabel('冰雹概率', fontsize=11)
    ax3.set_ylabel('频次', fontsize=11)
    ax3.set_title('冰雹概率分布', fontsize=12)
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # 典型阈值表格
    ax4 = axes[1, 1]
    ax4.axis('off')
    table_data = [
        ['参数', '冰雹特征', '阈值'],
        ['Z', '高值', '>50 dBZ'],
        ['ZDR', '高值', '>3.5 dB'],
        ['KDP', '高值', '>2 °/km'],
        ['ρhv', '低值', '<0.85']
    ]
    table = ax4.table(cellText=table_data, loc='center', cellLoc='center',
                      colWidths=[0.25, 0.35, 0.3])
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1.2, 1.8)
    # 表头加粗
    for i in range(3):
        table[(0, i)].set_facecolor('lightgray')
    ax4.set_title('冰雹检测阈值', fontsize=12, pad=20)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n检测到 {sum(1 for c in colors if c=='red')} / {n} 个冰雹样本")
```

```python
plot_hail_detection()
```
