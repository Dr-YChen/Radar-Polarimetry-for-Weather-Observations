# 5.3 退极化比与交叉相关系数

**学习目标**
- 理解线性退极化比(LDR)的定义和物理意义
- 掌握共极化相关系数ρhv的物理含义
- 观察不同降水类型下LDR和ρhv的典型值

**参考文献**：Ryzhkov & Zrnic, Chapter 5, pp. 536-560

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Dropdown
import ipywidgets as widgets
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def calculate_ldr_rhohv(sigma_theta_deg=15, particle_type='rain'):
    """
    简化计算LDR和ρhv
    基于倾角分散程度
    """
    sigma_theta = np.radians(sigma_theta_deg)
    
    # 简化模型
    if particle_type == 'rain':
        # 小雨滴高度一致
        rho_hv = np.exp(-2 * sigma_theta**2)
        ldr_db = -25 + sigma_theta_deg * 0.3  # 近似
    elif particle_type == 'snow':
        # 雪晶取向较随机
        rho_hv = np.exp(-3 * sigma_theta**2)
        ldr_db = -15 + sigma_theta_deg * 0.2
    elif particle_type == 'hail':
        # 冰雹形状不规则
        rho_hv = np.exp(-5 * sigma_theta**2) * 0.7
        ldr_db = -8 + sigma_theta_deg * 0.1
    else:
        rho_hv = 0.95
        ldr_db = -20
    
    return rho_hv, ldr_db

def plot_ldr_rhohv():
    """可视化LDR和ρhv"""
    sigma_range = np.linspace(1, 60, 100)
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # ρhv
    ax1 = axes[0]
    for ptype, color in [('rain', 'blue'), ('snow', 'green'), ('hail', 'red')]:
        rho_vals = []
        for sig in sigma_range:
            rho, _ = calculate_ldr_rhohv(sig, ptype)
            rho_vals.append(rho)
        ax1.plot(sigma_range, rho_vals, color=color, linewidth=2, label=ptype.capitalize())
    ax1.axhline(y=0.9, color='gray', linestyle='--', alpha=0.5, label='典型雨滴阈值')
    ax1.set_xlabel('倾角标准差 σθ (°)', fontsize=11)
    ax1.set_ylabel('|ρhv|', fontsize=11)
    ax1.set_title('共极化相关系数 |ρhv|', fontsize=12)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim(0.5, 1.05)
    
    # LDR
    ax2 = axes[1]
    for ptype, color in [('rain', 'blue'), ('snow', 'green'), ('hail', 'red')]:
        ldr_vals = []
        for sig in sigma_range:
            _, ldr = calculate_ldr_rhohv(sig, ptype)
            ldr_vals.append(ldr)
        ax2.plot(sigma_range, ldr_vals, color=color, linewidth=2, label=ptype.capitalize())
    ax2.set_xlabel('倾角标准差 σθ (°)', fontsize=11)
    ax2.set_ylabel('LDR (dB)', fontsize=11)
    ax2.set_title('线性退极化比 LDR', fontsize=12)
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== 典型极化特征值 ===")
    print(f"{'降水类型':<12} {'|ρhv|':<10} {'LDR (dB)':<15} {'特征':<20}")
    print("-" * 57)
    print(f"{'小雨':<12} {'~0.99':<10} {'<-25':<15} {'高度一致':<20}")
    print(f"{'大雨':<12} {'0.97-0.99':<10} {'-25 to -20':<15} {'轻微分散':<20}")
    print(f"{'雪':<12} {'0.90-0.98':<10} {'-25 to -15':<15} {'取向随机':<20}")
    print(f"{'冰雹':<12} {'0.70-0.90':<10} {'-15 to -5':<15} {'形状不规则':<20}")
```

```python
plot_ldr_rhohv()
```
