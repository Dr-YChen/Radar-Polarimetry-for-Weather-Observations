# 6.3 污染回波识别

**学习目标**
- 理解地物杂波的极化特征
- 掌握基于极化变量的杂波识别方法
- 了解噪声校正技术

**参考文献**：Ryzhkov & Zrnic, Chapter 6, pp. 631-660

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def plot_clutter_signature():
    """展示不同回波类型的极化特征"""
    echo_types = ['雨', '雪', '地物杂波', '海杂波', '生物飞虫']
    
    # 典型极化特征值（简化）
    Z_range = [30, 20, 50, 40, 15]  # dBZ
    ZDR_range = [2, 0.5, -0.5, 3, 4]  # dB
    rho_hv_range = [0.98, 0.95, 0.5, 0.7, 0.9]
    LDR_range = [-22, -18, -5, -10, -15]  # dB
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    colors = ['blue', 'cyan', 'brown', 'purple', 'green']
    
    # Z vs ZDR
    ax1 = axes[0, 0]
    for i, (etype, c) in enumerate(zip(echo_types, colors)):
        ax1.scatter(Z_range[i], ZDR_range[i], s=200, c=c, label=etype, zorder=5)
        ax1.annotate(etype, (Z_range[i], ZDR_range[i]), fontsize=9, ha='left')
    ax1.set_xlabel('Z (dBZ)', fontsize=11)
    ax1.set_ylabel('ZDR (dB)', fontsize=11)
    ax1.set_title('Z vs ZDR 特征空间', fontsize=12)
    ax1.grid(True, alpha=0.3)
    ax1.set_xlim(0, 60)
    ax1.set_ylim(-3, 6)
    
    # ρhv vs Z
    ax2 = axes[0, 1]
    for i, (etype, c) in enumerate(zip(echo_types, colors)):
        ax2.scatter(Z_range[i], rho_hv_range[i], s=200, c=c, zorder=5)
        ax2.annotate(etype, (Z_range[i], rho_hv_range[i]), fontsize=9)
    ax2.set_xlabel('Z (dBZ)', fontsize=11)
    ax2.set_ylabel('|ρhv|', fontsize=11)
    ax2.set_title('|ρhv| vs Z 特征空间', fontsize=12)
    ax2.grid(True, alpha=0.3)
    
    # 杂波识别阈值
    ax3 = axes[1, 0]
    categories = ['气象', '地物', '海杂波', '生物']
    rho_thresholds = [0.8, 0.6, 0.75, 0.85]
    colors_bar = ['lightblue', 'brown', 'purple', 'green']
    ax3.bar(categories, rho_thresholds, color=colors_bar, alpha=0.7, edgecolor='black')
    ax3.axhline(y=0.75, color='red', linestyle='--', label='杂波阈值 ρhv=0.75')
    ax3.set_ylabel('典型 |ρhv|', fontsize=11)
    ax3.set_title('不同回波类型的典型 |ρhv|', fontsize=12)
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # LDR分布
    ax4 = axes[1, 1]
    for i, (etype, c) in enumerate(zip(echo_types, colors)):
        ax4.scatter(LDR_range[i], Z_range[i], s=200, c=c, zorder=5)
        ax4.annotate(etype, (LDR_range[i], Z_range[i]), fontsize=9)
    ax4.axvline(x=-15, color='gray', linestyle=':', alpha=0.5)
    ax4.set_xlabel('LDR (dB)', fontsize=11)
    ax4.set_ylabel('Z (dBZ)', fontsize=11)
    ax4.set_title('LDR vs Z 特征空间', fontsize=12)
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== 杂波识别特征总结 ===")
    print("地物杂波: 高Z, 低ρhv (<0.7), 低ZDR")
    print("海杂波: 中Z, 低ρhv, 高ZDR")
    print("生物飞虫: 低Z, 中ρhv, 高ZDR")
    print("气象回波: ρhv > 0.9, ZDR与降水类型相关")
```

```python
plot_clutter_signature()
```
