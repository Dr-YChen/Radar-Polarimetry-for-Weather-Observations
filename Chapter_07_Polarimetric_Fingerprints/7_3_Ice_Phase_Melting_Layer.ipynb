# 7.3 冰相与融化层

**学习目标**
- 理解融化层的亮带效应
- 观察融化层的极化特征

**参考文献**：Ryzhkov & Zrnic, Chapter 7, pp. 731-760

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def melting_layer_profile(height_km, melting_layer_base=2.0, melting_layer_top=3.5):
    """
    模拟融化层的极化特征廓线
    """
    Z_profile = []
    ZDR_profile = []
    rho_hv_profile = []
    
    for h in height_km:
        if h < melting_layer_base:
            # 冰相区
            Z = 25 + (h - 1) * 5
            ZDR = 0.5
            rho = 0.95
        elif h > melting_layer_top:
            # 纯液相区
            Z = 35 + (h - melting_layer_top) * 3
            ZDR = 2.0 + (h - melting_layer_top) * 0.5
            rho = 0.98
        else:
            # 融化层
            melt_fraction = (h - melting_layer_base) / (melting_layer_top - melting_layer_base)
            # 亮带效应
            Z = 40 + 10 * np.sin(melt_fraction * np.pi)
            ZDR = 2.5 + 1.5 * melt_fraction
            rho = 0.85 - 0.1 * melt_fraction
        
        Z_profile.append(Z)
        ZDR_profile.append(ZDR)
        rho_hv_profile.append(rho)
    
    return np.array(Z_profile), np.array(ZDR_profile), np.array(rho_hv_profile)

def plot_melting_layer():
    """可视化融化层特征"""
    height = np.linspace(0.5, 5.0, 100)
    
    Z, ZDR, rho = melting_layer_profile(height)
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 6))
    
    # Z廓线
    ax1 = axes[0]
    ax1.plot(Z, height, 'b-', linewidth=2)
    ax1.fill_betweenx(height, Z, Z*0.7, alpha=0.3)
    ax1.axhline(y=2.0, color='green', linestyle='--', label='融化层底')
    ax1.axhline(y=3.5, color='red', linestyle='--', label='融化层顶')
    ax1.annotate('亮带', xy=(Z.max()*0.9, 2.75), fontsize=10, color='red')
    ax1.set_xlabel('Z (dBZ)', fontsize=11)
    ax1.set_ylabel('高度 (km)', fontsize=11)
    ax1.set_title('反射率廓线 - 亮带效应', fontsize=12)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    ax1.set_xlim(20, 55)
    
    # ZDR廓线
    ax2 = axes[1]
    ax2.plot(ZDR, height, 'g-', linewidth=2)
    ax2.axhline(y=2.0, color='green', linestyle='--')
    ax2.axhline(y=3.5, color='red', linestyle='--')
    ax2.set_xlabel('ZDR (dB)', fontsize=11)
    ax2.set_ylabel('高度 (km)', fontsize=11)
    ax2.set_title('差分反射率廓线', fontsize=12)
    ax2.grid(True, alpha=0.3)
    
    # ρhv廓线
    ax3 = axes[2]
    ax3.plot(rho, height, 'r-', linewidth=2)
    ax3.axhline(y=2.0, color='green', linestyle='--')
    ax3.axhline(y=3.5, color='red', linestyle='--')
    ax3.set_xlabel('|ρhv|', fontsize=11)
    ax3.set_ylabel('高度 (km)', fontsize=11)
    ax3.set_title('相关系数廓线', fontsize=12)
    ax3.grid(True, alpha=0.3)
    ax3.set_xlim(0.8, 1.0)
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== 融化层极化指纹 ===")
    print("亮带特征:")
    print("  - Z 在融化层出现峰值（+10dB）")
    print("  - ZDR 在融化层升高")
    print("  - ρhv 在融化层降低（粒子形状不均匀）")
    print("应用: 识别0°C层高度，估算融化层厚度")
```

```python
plot_melting_layer()
```
