# 7.2 尺寸分选效应

**学习目标**
- 理解风切变导致的尺寸分选机制
- 观察分选效应对极化变量的影响

**参考文献**：Ryzhkov & Zrnic, Chapter 7, pp. 701-730

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def size_sorting_effect(wind_shear=20, distance_from_shear=0):
    """
    模拟尺寸分选效应
    wind_shear: 风切变强度 (m/s/km)
    distance: 距风切变层的距离 (km)
    """
    # 速度差导致粒子分离
    separation = wind_shear * distance_from_shear * 0.1
    
    # 近处：粗粒子（大雨滴）更多（沉降快）
    # 远处：细粒子（小雨滴）更多
    D0_near = 2.0 + separation * 0.05  # mm
    D0_far = 2.0 - separation * 0.03  # mm
    
    D0 = np.clip(D0_near if distance_from_shear < 5 else D0_far, 0.5, 4.0)
    
    # ZDR也受影响
    ZDR = 1.5 + separation * 0.1
    ZDR = np.clip(ZDR, 0, 5)
    
    return D0, ZDR

def plot_size_sorting():
    """可视化尺寸分选效应"""
    distances = np.linspace(0, 20, 100)
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # D0随距离变化
    ax1 = axes[0, 0]
    for shear in [10, 20, 30]:
        D0_vals = [size_sorting_effect(shear, d)[0] for d in distances]
        ax1.plot(distances, D0_vals, linewidth=2, label=f'风切变={shear} m/s/km')
    ax1.set_xlabel('距风切变层距离 (km)', fontsize=11)
    ax1.set_ylabel('D0 (mm)', fontsize=11)
    ax1.set_title('尺寸分选导致D0变化', fontsize=12)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # ZDR随距离变化
    ax2 = axes[0, 1]
    for shear in [10, 20, 30]:
        ZDR_vals = [size_sorting_effect(shear, d)[1] for d in distances]
        ax2.plot(distances, ZDR_vals, linewidth=2, label=f'风切变={shear} m/s/km')
    ax2.set_xlabel('距风切变层距离 (km)', fontsize=11)
    ax2.set_ylabel('ZDR (dB)', fontsize=11)
    ax2.set_title('尺寸分选导致ZDR变化', fontsize=12)
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 2D示意
    ax3 = axes[1, 0]
    # 风切变层上方：大滴被带走
    # 风切变层下方：大滴沉降快
    ax3.annotate('', xy=(10, 8), xytext=(10, 2),
                arrowprops=dict(arrowstyle='->', color='blue', lw=2))
    ax3.text(11, 5, '风切变', fontsize=10)
    ax3.fill_between([0, 20], [5, 5], [10, 10], alpha=0.2, color='blue', label='大风速区')
    ax3.fill_between([0, 20], [0, 0], [5, 5], alpha=0.1, color='green', label='小风速区')
    ax3.scatter([5, 5], [8, 3], s=[200, 100], c=['red', 'blue'], alpha=0.7, label='大雨滴/小滴')
    ax3.set_xlabel('水平距离 (km)', fontsize=11)
    ax3.set_ylabel('高度 (km)', fontsize=11)
    ax3.set_title('尺寸分选示意图', fontsize=12)
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # 典型极化特征
    ax4 = axes[1, 1]
    ax4.axis('off')
    text = """
    尺寸分选的极化指纹：
    
    特征：
    - ZDR 在风切变层附近出现峰值
    - D0 在上风方向较大
    - KDP 可能有类似ZDR的分选特征
    
    应用：
    - 识别对流风暴中的上升气流区
    - 估计风切变强度
    - 分析降水粒子的传输路径
    """
    ax4.text(0.1, 0.5, text, fontsize=11, verticalalignment='center',
            family='monospace', bbox=dict(boxstyle='round', facecolor='lightyellow'))
    ax4.set_title('尺寸分选极化特征', fontsize=12)
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== 尺寸分选极化指纹 ===")
    print("风切变导致不同大小的粒子在不同位置聚集")
    print("大雨滴落速快，在下沉过程中水平传输更远")
    print("导致ZDR在风切变区呈现特征性分布")
```

```python
plot_size_sorting()
```
