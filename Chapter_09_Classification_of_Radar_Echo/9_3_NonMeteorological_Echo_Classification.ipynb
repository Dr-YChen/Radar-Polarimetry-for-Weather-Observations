# 9.3 非气象目标识别

**参考文献**：Ryzhkov & Zrnic, Chapter 9, pp. 931-960

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def plot_nonmeteoro_echo():
    """非气象目标极化特征"""
    echo_types = ['地物', '海杂波', '生物', '沙尘', '降水']
    
    # 典型值
    Z_range = [50, 40, 15, 30, 35]
    ZDR_range = [-0.5, 3, 4, 2, 2]
    rho_range = [0.5, 0.7, 0.85, 0.8, 0.97]
    LDR_range = [-5, -10, -12, -8, -22]
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    colors = ['brown', 'purple', 'green', 'orange', 'blue']
    
    # Z vs ZDR
    ax1 = axes[0, 0]
    for i, (etype, c) in enumerate(zip(echo_types, colors)):
        ax1.scatter(Z_range[i], ZDR_range[i], s=200, c=c, label=etype)
    ax1.set_xlabel('Z (dBZ)', fontsize=11)
    ax1.set_ylabel('ZDR (dB)', fontsize=11)
    ax1.set_title('Z vs ZDR', fontsize=12)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # ρhv vs LDR
    ax2 = axes[0, 1]
    for i, (etype, c) in enumerate(zip(echo_types, colors)):
        ax2.scatter(rho_range[i], LDR_range[i], s=200, c=c)
        ax2.annotate(etype, (rho_range[i], LDR_range[i]), fontsize=9)
    ax2.set_xlabel('|ρhv|', fontsize=11)
    ax2.set_ylabel('LDR (dB)', fontsize=11)
    ax2.set_title('|ρhv| vs LDR', fontsize=12)
    ax2.grid(True, alpha=0.3)
    
    # 分类决策树
    ax3 = axes[1, 0]
    ax3.axis('off')
    text = """
    非气象目标识别流程：
    
    1. ρhv < 0.8 → 可能非气象
    2. 若 ρhv > 0.8 → 很可能是气象
    
    3. 地物：低ZDR, 极低ρhv
    4. 海杂波：高ZDR, 低ρhv
    5. 生物：高ZDR, 中等ρhv
    6. 沙尘：中ZDR, 低ρhv
    """
    ax3.text(0.1, 0.5, text, fontsize=11, verticalalignment='center',
             family='monospace', bbox=dict(boxstyle='round', facecolor='lightyellow'))
    ax3.set_title('非气象目标识别', fontsize=12)
    
    # 特征对比表
    ax4 = axes[1, 1]
    ax4.axis('off')
    table_data = [
        ['类型', 'Z', 'ZDR', 'ρhv', 'LDR'],
        ['地物', '高', '低', '很低', '高'],
        ['海杂波', '中', '高', '低', '中'],
        ['生物', '低', '高', '中', '中'],
        ['沙尘', '中', '中', '低', '中'],
        ['降水', '中', '中', '高', '低']
    ]
    table = ax4.table(cellText=table_data, loc='center', cellLoc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1.2, 1.8)
    for i in range(5):
        table[(0, i)].set_facecolor('lightgray')
    ax4.set_title('极化特征对比表', fontsize=12, pad=20)
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== 非气象目标识别 ===")
    print("关键是组合多个极化变量")
    print("单一变量不足以区分所有类型")
    print("需要结合空间分布和时间变化")
```

```python
plot_nonmeteoro_echo()
```
