# 11.2 冰相粒子微物理反演

**参考文献**：Ryzhkov & Zrnic, Chapter 11, pp. 1101-1140

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def retrieve_ice_microphysics(Z, ZDR, KDP):
    """
    从 Z, ZDR, KDP 反演冰相微物理参数
    - IWC: 冰水含量
    - Dm: 质量加权平均直径
    """
    # 冰的介电特性
    K_ice = 0.176 / 0.93  # 冰的|K|²
    K_water = 0.93  # 水的|K|²
    
    # IWC 估算（经验公式）
    # IWC 与 Z 的关系取决于粒子类型
    Z_linear = 10**(Z/10)
    
    # 对于雪花聚合体
    IWC = 0.004 * Z_linear**0.4  # g/m³
    
    # Dm 估算
    Dm = 2.0 + 0.5 * ZDR  # mm
    
    return IWC, Dm

def plot_ice_retrieval():
    """冰相反演示意"""
    Z_range = np.linspace(15, 40, 100)
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # IWC-Z 关系
    ax1 = axes[0, 0]
    IWC = 0.004 * (10**(Z_range/10))**0.4
    ax1.semilogy(Z_range, IWC, 'c-', linewidth=2)
    ax1.set_xlabel('Z (dBZ)', fontsize=11)
    ax1.set_ylabel('IWC (g/m³)', fontsize=11)
    ax1.set_title('IWC-Z 关系 (雪花)', fontsize=12)
    ax1.grid(True, alpha=0.3)
    
    # 不同冰粒子类型的对比
    ax2 = axes[0, 1]
    ice_types = ['柱状冰晶', '板状冰晶', '雪花聚合体', '霰粒']
    K_values = [0.15, 0.18, 0.12, 0.10]
    colors = ['cyan', 'lightblue', 'white', 'gray']
    bars = ax2.bar(ice_types, K_values, color=colors, edgecolor='black')
    ax2.set_ylabel('雷达反射效率 K²', fontsize=11)
    ax2.set_title('不同冰粒子类型的雷达特性', fontsize=12)
    ax2.tick_params(axis='x', rotation=15)
    
    # 反演不确定性
    ax3 = axes[1, 0]
    ax3.axis('off')
    text = """
    冰相反演的挑战：
    
    1. 冰的介电常数远小于水
    2. 粒子形状复杂多样
    3. 融化过程混合相态
    4. 粒子取向分布不确定
    
    精度：
    - IWC: 50-70% 误差
    - Dm: 30-40% 误差
    """
    ax3.text(0.1, 0.5, text, fontsize=11, verticalalignment='center',
             family='monospace', bbox=dict(boxstyle='round', facecolor='lightyellow'))
    ax3.set_title('冰相反演的挑战', fontsize=12)
    
    # 高度-冰水含量关系
    ax4 = axes[1, 1]
    height = np.linspace(3, 10, 100)
    IWC_height = 0.5 * np.exp(-height / 5)  # 简化模型
    ax4.plot(IWC_height, height, 'b-', linewidth=2)
    ax4.fill_betweenx(height, IWC_height, alpha=0.3)
    ax4.set_xlabel('IWC (g/m³)', fontsize=11)
    ax4.set_ylabel('高度 (km)', fontsize=11)
    ax4.set_title('IWC 垂直分布', fontsize=12)
    ax4.grid(True, alpha=0.3)
    ax4.invert_yaxis()
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== 冰相反演 ===")
    print("精度低于雨相反演")
    print("需要结合温度廓线")
    print("多参数融合可改善精度")
```

```python
plot_ice_retrieval()
```
