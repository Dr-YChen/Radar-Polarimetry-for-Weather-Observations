# 11.1 纯雨条件下的微物理反演

**参考文献**：Ryzhkov & Zrnic, Chapter 11, pp. 1061-1100

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def retrieve_rain_microphysics(Z, ZDR, KDP):
    """
    从 Z, ZDR, KDP 反演雨滴微物理参数
    - D0: 中值直径
    - N0*: 截距参数
    - LWC: 液态水含量
    """
    # 简化的反演关系
    # D0 与 ZDR 相关
    D0 = 0.5 + 0.35 * ZDR  # mm
    
    # N0* 与 Z 和 D0 相关
    Z_linear = 10**(Z/10)
    N0_star = Z_linear / (D0**6) * 0.01  # 简化公式
    
    # LWC 估算
    LWC = 0.0006 * Z_linear**0.5 * (1 + ZDR/3)  # g/m³
    
    return D0, N0_star, LWC

def plot_retrieval():
    """反演示意"""
    ZDR_range = np.linspace(0.5, 4, 100)
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # D0 vs ZDR
    ax1 = axes[0, 0]
    D0 = 0.5 + 0.35 * ZDR_range
    ax1.plot(ZDR_range, D0, 'b-', linewidth=2)
    ax1.set_xlabel('ZDR (dB)', fontsize=11)
    ax1.set_ylabel('D0 (mm)', fontsize=11)
    ax1.set_title('D0-ZDR 关系', fontsize=12)
    ax1.grid(True, alpha=0.3)
    
    # LWC vs Z
    ax2 = axes[0, 1]
    Z_range = np.linspace(20, 50, 100)
    LWC = 0.0006 * 10**(Z_range/10)**0.5 * 1.5  # 假设ZDR=2
    ax2.semilogy(Z_range, LWC, 'g-', linewidth=2)
    ax2.set_xlabel('Z (dBZ)', fontsize=11)
    ax2.set_ylabel('LWC (g/m³)', fontsize=11)
    ax2.set_title('LWC-Z 关系', fontsize=12)
    ax2.grid(True, alpha=0.3)
    
    # 反演流程
    ax3 = axes[1, 0]
    ax3.axis('off')
    text = """
    微物理反演流程：
    
    输入: Z, ZDR, KDP
    
    1. D0 = f(ZDR)
       - ZDR↑ → D0↑ (大滴更多)
    
    2. N0* = g(Z, D0)
       - 调节截距参数
    
    3. LWC = h(Z, ZDR)
       - 综合考虑粒子浓度和大小
    """
    ax3.text(0.1, 0.5, text, fontsize=11, verticalalignment='center',
             family='monospace', bbox=dict(boxstyle='round', facecolor='lightyellow'))
    ax3.set_title('反演算法流程', fontsize=12)
    
    # 误差敏感性
    ax4 = axes[1, 1]
    params = ['D0', 'N0*', 'LWC']
    sensitivity = [15, 40, 25]  # %误差
    colors = ['blue', 'red', 'green']
    bars = ax4.bar(params, sensitivity, color=colors, alpha=0.7)
    ax4.set_ylabel('反演误差 (%)', fontsize=11)
    ax4.set_title('微物理参数反演误差', fontsize=12)
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== 雨微物理反演 ===")
    print("D0 反演最准确 (~15%误差)")
    print("N0* 离散度大 (~40%误差)")
    print("LWC 中等精度 (~25%误差)")
```

```python
plot_retrieval()
```
