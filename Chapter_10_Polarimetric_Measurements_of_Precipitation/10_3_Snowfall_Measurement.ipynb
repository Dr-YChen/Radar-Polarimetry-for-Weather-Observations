# 10.3 降雪量测量

**参考文献**：Ryzhkov & Zrnic, Chapter 10, pp. 1031-1060

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def plot_snowfall_estimation():
    """降雪估计方法"""
    # 雪密度范围
    rho_snow = np.linspace(50, 300, 100)  # kg/m^3
    
    # 雪水当量系数 K
    K = 1000 / rho_snow
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    ax1 = axes[0]
    ax1.plot(rho_snow, K, 'b-', linewidth=2)
    ax1.set_xlabel('雪密度 (kg/m³)', fontsize=11)
    ax1.set_ylabel('K = SWE/Z 系数', fontsize=11)
    ax1.set_title('雪水当量系数与密度关系', fontsize=12)
    ax1.grid(True, alpha=0.3)
    
    ax2 = axes[1]
    snow_types = ['新雪', '细雪', '粗雪', '湿雪']
    densities = [50, 100, 200, 300]
    colors = ['white', 'lightgray', 'gray', 'darkgray']
    bars = ax2.bar(snow_types, densities, color=colors, edgecolor='black')
    ax2.set_ylabel('雪密度 (kg/m³)', fontsize=11)
    ax2.set_title('不同类型雪的典型密度', fontsize=12)
    ax2.grid(True, alpha=0.3)
    for bar, val in zip(bars, densities):
        ax2.text(bar.get_x() + bar.get_width()/2, val + 5, f'{val}', ha='center')
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== 降雪估计 ===")
    print("SWE = K * Z^0.5")
    print("K 与雪密度成反比")
    print("新雪 K ≈ 15-20")
    print("老雪/湿雪 K ≈ 5-8")
```

```python
plot_snowfall_estimation()
```
