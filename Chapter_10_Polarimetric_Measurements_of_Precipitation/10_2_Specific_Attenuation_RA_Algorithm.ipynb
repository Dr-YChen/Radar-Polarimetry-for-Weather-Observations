# 10.2 R(A) 比衰减算法

**参考文献**：Ryzhkov & Zrnic, Chapter 10, pp. 1001-1030

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def RA_algorithm(A, wavelength_mm):
    """
    R(A) 比衰减算法
    A: 比衰减 dB/km
    """
    if wavelength_mm < 4:  # X-band
        return 12.6 * A**0.93
    elif wavelength_mm < 8:  # C-band
        return 9.2 * A**0.95
    else:  # S-band
        return 6.8 * A**1.0

def plot_RA_advantage():
    """展示R(A)算法的优势"""
    A_range = np.linspace(0.1, 3, 100)
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # R(A) 关系
    ax1 = axes[0]
    for band, wl in [('X', 3.2), ('C', 5.4), ('S', 10.7)]:
        R = [RA_algorithm(a, wl) for a in A_range]
        ax1.plot(A_range, R, linewidth=2, label=f'{band}-band')
    ax1.set_xlabel('比衰减 A (dB/km)', fontsize=11)
    ax1.set_ylabel('降雨率 R (mm/h)', fontsize=11)
    ax1.set_title('R(A) 降雨估计', fontsize=12)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # R(A)的优势
    ax2 = axes[1]
    advantages = ['不受衰减影响', '不受定标偏差影响', '对小雨更敏感', '空间一致性']
    scores = [95, 90, 85, 80]
    colors = ['#FFD700', '#C0C0C0', '#CD7F32', '#4682B4']
    bars = ax2.barh(advantages, scores, color=colors, alpha=0.7)
    ax2.set_xlabel('相对优势评分', fontsize=11)
    ax2.set_title('R(A)算法的优势', fontsize=12)
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== R(A) 算法优势 ===")
    print("A (比衰减) 可以从 PhiDP 直接估算")
    print("不依赖绝对反射率定标")
    print("在X波段特别有效")
    print("对小雨估计比R(Z)更可靠")
```

```python
plot_RA_advantage()
```
