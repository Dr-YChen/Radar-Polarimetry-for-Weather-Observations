# 10.1 降雨率估计模型对比

**参考文献**：Ryzhkov & Zrnic, Chapter 10, pp. 961-1000

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['DejaVu Sans', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def R_Z(Z_dbz):
    """传统 R-Z 关系"""
    Z = 10**(Z_dbz/10)
    return (Z / 200)**(1/1.5)

def R_ZDR(ZDR_db):
    """R-ZDR 关系"""
    return 0.039 * 10**(ZDR_db / 10)

def R_KDP(KDP):
    """R-KDP 关系 (S-band)"""
    return 44.0 * KDP**0.822

def plot_rainfall_comparison():
    """对比不同降雨估计方法"""
    Z_range = np.linspace(20, 55, 100)
    ZDR_range = np.linspace(0.5, 4.5, 100)
    KDP_range = np.linspace(0.1, 5, 100)
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # R-Z 关系
    ax1 = axes[0, 0]
    R_z = [R_Z(z) for z in Z_range]
    ax1.semilogy(Z_range, R_z, 'b-', linewidth=2)
    ax1.set_xlabel('Z (dBZ)', fontsize=11)
    ax1.set_ylabel('R (mm/h)', fontsize=11)
    ax1.set_title('R-Z 关系: R=(Z/200)^0.67', fontsize=12)
    ax1.grid(True, alpha=0.3)
    
    # R-ZDR 关系
    ax2 = axes[0, 1]
    R_zdr = [R_ZDR(zdr) for zdr in ZDR_range]
    ax2.plot(ZDR_range, R_zdr, 'g-', linewidth=2)
    ax2.set_xlabel('ZDR (dB)', fontsize=11)
    ax2.set_ylabel('R (mm/h)', fontsize=11)
    ax2.set_title('R-ZDR 关系', fontsize=12)
    ax2.grid(True, alpha=0.3)
    
    # R-KDP 关系
    ax3 = axes[1, 0]
    R_kdp = [R_KDP(k) for k in KDP_range]
    ax3.plot(KDP_range, R_kdp, 'r-', linewidth=2)
    ax3.set_xlabel('KDP (°/km)', fontsize=11)
    ax3.set_ylabel('R (mm/h)', fontsize=11)
    ax3.set_title('R-KDP 关系 (S-band)', fontsize=12)
    ax3.grid(True, alpha=0.3)
    
    # 误差对比
    ax4 = axes[1, 1]
    # 模拟R-Z的离散度
    true_R = 20  # mm/h
    Z_at_R = 10*np.log10(200 * true_R**1.5)
    Z_measured = Z_at_R + np.random.randn(100) * 5
    R_from_Z = [R_Z(z) for z in Z_measured]
    
    methods = ['R(Z)', 'R(ZDR)', 'R(KDP)']
    errors = [np.std(R_from_Z), 3, 2]  # 典型误差
    colors = ['blue', 'green', 'red']
    bars = ax4.bar(methods, errors, color=colors, alpha=0.7)
    ax4.set_ylabel('估计误差 (%)', fontsize=11)
    ax4.set_title('降雨估计方法精度对比', fontsize=12)
    ax4.grid(True, alpha=0.3)
    for bar, val in zip(bars, errors):
        ax4.text(bar.get_x() + bar.get_width()/2, val + 0.5, f'{val:.0f}%', ha='center')
    
    plt.tight_layout()
    plt.show()
    
    print("\n=== 降雨估计方法对比 ===")
    print("R(Z): 经典方法，但离散度大 (~40-50%误差)")
    print("R(ZDR): 中等精度，受定标误差影响")
    print("R(KDP): 最稳定，不受衰减影响")
    print("多参数融合可进一步提高精度")
```

```python
plot_rainfall_comparison()
```
